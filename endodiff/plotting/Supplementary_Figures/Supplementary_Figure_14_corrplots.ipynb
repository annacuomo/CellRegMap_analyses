```r
library(corrplot)
```

```r
C1_file = "/hps/nobackup2/stegle/users/acuomo/all_scripts/struct_LMM2/sc_endodiff/debug_May2021/mofa_logcounts_model_factors.csv"
C2_file = "/hps/nobackup2/stegle/users/acuomo/all_scripts/struct_LMM2/sc_endodiff/new/input_files/20PCs.csv"
C3_file = "/hps/nobackup2/stegle/users/acuomo/CellRegMap_data/ldvae_emb.csv"
C4_file = "/hps/nobackup2/stegle/users/acuomo/CellRegMap_data/zinbwave_20.csv"
```

```r
df_m = read.csv(C1_file, row.names=1)
head(df_m)
dim(df_m)
```

```r
df_p = read.csv(C2_file, row.names=1)
head(df_p)
dim(df_p)
```

```r
######## MOFA vs PCA
```

```r
df = cbind(df_m[,1:10], df_p[,1:10])
```

```r
options(repr.plot.width = 5, repr.plot.height = 5)
corrplot(cor(df))
```

```r
fig_dir = "/hps/nobackup2/stegle/users/acuomo/CellRegMap_data/figures/"
pdf(paste0(fig_dir,"SF10_corrplot_MOFA_PCA.pdf"), width=5, height=5)
corrplot(cor(df))
dev.off()
```
